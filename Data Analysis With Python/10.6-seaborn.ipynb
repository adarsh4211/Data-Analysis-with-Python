#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots -- they uses categorical variables ( i.e, with categories)
## BAr Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
## Box Plot  - Amazing way to probably see the outliers
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
## Violin Plot  - use to show distribution of particular data (i.e, mean and all)

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)  # kde(kernel Density estimater) gets the shape of the histogram using a curve
```

```python
## KDE Plot -- to show only the curve
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot  - Here we will see the relationship between each and every feature
sns.pairplot(tips)
```

```python
## HEatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')  # annotation = true  , cmap(colour map) -- we specify colour using this
# Red -- Depicts the most coorelation between 2 features
```

```python
import pandas as pd
sales_df=pd.read_csv('data.csv')
sales_df.head()
```

```python
## Plot total sales by category
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y="Sales",data=sales_df,estimator=sum) #estimaror -- in  matplotlib we need to apply aggregator function, but here we can use this estimator feature and inside this can put any aggregation function
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product',y="Sales",data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

### Reading Data From Different Sources


```python
import pandas as pd
from io import StringIO
#Data in json format(i.e, key value pairs, similar to dictonary of python expect that it only accepts strings in double quotes"")
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
#we are converting json into string using StringIO and then into a dataframe
df=pd.read_json(StringIO(Data))
```

```python
df
```

```python
df.to_json()  # to convert dataframe back to json
# we have not given any parameters, so it will print index, value for each key value pairs
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",header=None)
```

```python
df.head()  # As we did header=None, so it will show default column names with 0-13
```

```python
df.to_csv("wine.csv") # we converted the above dataframe back to csv file and stored in a new csv file named wine.csv
```
